```python
# Import Libraries

import torch
import os
import json
import shutil
from ultralytics.nn.tasks import DetectionModel
```

```python
with torch.serialization.safe_globals([DetectionModel]):
    from ultralytics import YOLO
    model = YOLO('yolov8s.pt')

    # Train on custom dataset described in data.yaml
    model.train(data='configs/data.yaml', epochs=50, imgsz=640, batch=16)

    # Evaluate and save as needed
    metrics = model.val()
    save_dir = 'runs/train/motorbike_yolov8s/weights'
    os.makedirs(save_dir, exist_ok=True)

    # Save model weights
    model.save(os.path.join(save_dir, 'best.pt'))

    # Save evaluation metrics
    with open(os.path.join(save_dir, 'evaluation_metrics.json'), 'w') as f:
        json.dump(str(metrics), f, indent=4)

    # Copy validation images and plots folder
    val_results_dir = 'runs/val'
    dest_dir = os.path.join(save_dir, 'val_results_images')
    if os.path.exists(dest_dir):
        shutil.rmtree(dest_dir)
    if os.path.exists(val_results_dir):
        shutil.copytree(val_results_dir, dest_dir)

    print(f"Model, metrics, and validation images saved to {save_dir}")
```
